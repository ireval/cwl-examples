```python
# Import the metrics and plotting packages
from helpers import *
%matplotlib inline

# Define the topics 
# For each topic, provide a gain vector, and a cost vector

# Topic One
gains1 = [ 0,0,0.2,0.4,1.0,0.2,0,0,1.0,0,0,0.4,0,0,0]
costs1 = [ 1.2, 0.6, 0.4, 0.6, 3.6, 1.6, 0.6, 2.6, 0.6, 0.6, 0.6, 0.6, 0.6, 0.6, 1.8]
topic1 = Ranking("T1",gains1,costs1)

# Topic Two
gains2 = [ 1,0,1,0.4,0.0,0.2,0,0,1.0,0.2,0,0.4,0,0,0]
costs2 = [ 3.2, 1.6, 1.4, 0.6, 3.6, 1.6, 0.6, 1.6, 2.6, 0.2, 1.2, 0.2, 0.2, 0.6, 1.8]
topic2 = Ranking("T2",gains2,costs2)

topics = [topic1, topic2]
```

```python
#Create the Metric and Plot the C W and L
rbp = RBPCWLMetric(theta=0.6)
plot_cwl_figures(rbp, topics)

rbp.measure(topic1)
# The cwl_eval program output the following for each topic
# TOPIC ID   METRIC   EU  ETU   EC   ETC    ED
rbp.report()

rbp.measure(topic2)
# The cwl_eval program output the following for each topic
rbp.report()

# Try varying the parameters for each of the metrics to see 
# how the continuation, weights and likelihood of stopping changes

# RBP's, NDCG's, and Precision's C W and L functions are the same for a given theta - 
# and do not change depending on the gains and costs.

# But when AP, TBG, IFT and BPM are used - then the C W and L functions change 
# depending on the the gains and costs for the topics
```

```python
# Precision
precision = PrecisionCWLMetric(5)
plot_cwl_figures(precision, topics)
precision.measure(topic1)
precision.report()
precision.measure(topic2)
precision.report()
```
